```python
import import_ipynb
from search import *
import cv2
import matplotlib.pyplot as plt
import numpy as np
import copy


# 하나씩 잘려진 숫자 이미지를 배열 형태로 반환
def get_numbers_img(numbers):
#     numbers = search(link)
#     numbers = numbers[1:]

    ls = [[] for _ in range(100)]

    cnt = 0
    index = numbers[0][-9:-5]

    for data in numbers:
        if len(ls[0]) == 0:
            ls[cnt].append(data)
        else:
            if index == data[-9:-5]:
                ls[cnt].append(data)
            else:
                index = data[-9:-5]
                cnt += 1
                ls[cnt].append(data)

    ls = list(filter(None, ls))
        
    return ls

# 하나씩 잘려진 숫자 이미지를 배열로 받아 바이너리 이미지로 변환하여 저장
def get_binary_numbers(ls):
    # 바이너리 이미지를 저장할 배열을 ls와 똑같은 크기로 선언하기
    binary_numbers = copy.deepcopy(ls)
    
    index = 0
    cnt = 0
    
    # ls 내의 컬러 이미지를 바이너리 형태로 변환하여 binary_numbers 배열에 저장
    for data in ls:
        for link in data:
            image = cv2.imread(link)
            image = cv2.resize(image,(12,12))

            img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

            _, img_thresh = cv2.threshold(img_gray, 150, 255, cv2.THRESH_BINARY)
            out = 255-img_thresh

            binary_numbers[index][cnt] = out
            cnt += 1

        index += 1
        cnt = 0

    # 파일 이름을 동일하게 저장하기 위한 배열 -> ls를 1차원 배열로 변환
    fnames = []
    index = 0
    for data in ls:
        for link in data:
            fnames.append(link)

    # binary_numbers 에 들어있는 이미지를 mnist 데이터와 호환할 수 있게 20*20 이미지 형태로 변환 후 저장
    for data_arr in binary_numbers:
        for data in data_arr:
            bg = np.full((20,20), 255)

            for i in range(12):
                for j in range(12):
                    bg[i+4][j+4] = data[i][j]


            fname = './binary_numbers/' + fnames[index][10:]

            cv2.imwrite(fname, bg)

            index += 1


# numbers_list = get_numbers_img('./numbers')
# get_binary_numbers(numbers_list)
```
